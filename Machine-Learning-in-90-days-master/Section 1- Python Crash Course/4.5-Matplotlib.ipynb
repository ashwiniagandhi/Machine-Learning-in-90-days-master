## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)#2 rows 1 clm
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) #bins value are by default is 10
plt.title("histogram") 
plt.show()#use for specify count and density
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]# use normal distribution between  lower value means 0 to standard distrubuution value is std,taking step size by 100
#for loop are  1 to 4 range.1st is 0 to std is taking 1 ,next loop is 0 to 2 put 2 bcoz range between 1 to 4.
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  #3 box plot bcoz loop iterate 1 to 4 means total 3 box created.
#box plot helps to find out percentile.circle shows outlier
```

```python
plt.boxplot(data,vert=True,patch_artist=True);#if we write patch_artist is True color is shown if patch artist falsethen is goes off
```

```python
plt.boxplot(data,vert=False,patch_artist=False);
```

```python
plt.boxplot(data,vert=True,patch_artist=True);
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]#if calculate tatal value is 800#each every value get percentage.215 means itmay to 20 to 25%#pie char calculate/shows  on percentage 
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice#python are splited

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```

```python
explode = (0.4, 0.2, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)#autopct use for format.wha format you have.1.1 f means you have floting format.shadow true means shadow shows

plt.axis('equal')
plt.show()
```
